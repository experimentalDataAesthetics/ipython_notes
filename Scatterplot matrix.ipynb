## How to automatize data reading, scaling and scatterplot matrix?

This is the sequence of code from the example above to loop and simply

```python
import numpy as np
from StringIO import StringIO
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.gridspec as gridspec
```

```python
data = np.genfromtxt("iris.data", dtype=float, delimiter=',', skip_header=1) 
data_scaled = (data - np.min(data, axis=0))/(np.max(data, axis=0) - np.min(data, axis=0))
```

```python
data_scaled.shape
```

```python
def scatterplotMatrix ( sizeinch, dpival, dimensions, filename ):
    fig = plt.figure(figsize=(sizeinch, sizeinch), dpi=dpival)
    
    matrixlength = dimensions-1 # 4-1 = 3

    gs = gridspec.GridSpec(matrixlength, matrixlength)
    
    #here comes a loop creating the matrix
    for i in range(0, matrixlength): 
        for ii in range(0, matrixlength): 
            plt.subplot(gs[i, ii]).scatter(data_scaled[:,(ii+1)],data_scaled[:,i], alpha=0.5, c='#262626', linewidth=0.15)    
    fig.savefig(filename)
```

```python
scatterplotMatrix (15, 400, 4, 'scatterplotmatrix_iris.pdf')
```
