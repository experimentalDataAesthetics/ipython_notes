# EXPLORATIONS WITH ARRAY READING USING OSCGRAIN

### Use code below in SC3 to load the Synthdefs, that create the grain sounds

```python
import numpy as np
from StringIO import StringIO
from IPython.display import Image
import OSC
import time, sys
import time, random
client = OSC.OSCClient()
client.connect( ( '127.0.0.1', 57110 ) )
```

```python
def oscgrain( frequency ):
    msg = OSC.OSCMessage()
    msg.setAddress("s_new")
    msg.append("grain")
    msg.append(-1)
    msg.append(0)
    msg.append(1)
    msg.append("amp")
    msg.append(1)
    msg.append("freq")
    msg.append(frequency)     #read in data points
    msg.append("sustain")
    msg.append(0.015) #0.01-0.04
    msg.append("pan")
    msg.append(-1)
    client.send(msg)
```

Playing a grain with frequency 1000Hz

```python
oscgrain(1000)
```

### Min Max Scale all data values to values between 0 and 1

```python
data_all = (data - np.min(data, axis=0))/(np.max(data, axis=0) - np.min(data, axis=0))
```

```python
data_all
```

### read-in one data point

```python
for i in range(4):
    oscgrain(2000+(1500*data_all[1,i]))
    time.sleep(0.9)
```

### read-in all data point sequentially as listed in the file

```python
for i in range(149):
    for ii in range (4):
        oscgrain(2000+(1500*data_all[i,ii]))
        time.sleep(0.01) # 14*0.01=0.14 lenght/period time
    time.sleep(0.2)
```

Use data.shape for loop settings

## and play

```python
for i in range(np.asarray(data.shape)[0]):
    for ii in range (np.asarray(data.shape)[1]-1): #last dimension is not a float = class information
        oscgrain(2000+(1500*data_all[i,ii]))
        time.sleep(0.01) # 14*0.01=0.14 lenght/period time
    time.sleep(0.2)
```
