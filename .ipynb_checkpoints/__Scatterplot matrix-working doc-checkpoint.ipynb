## Scatterplot matrix

http://matplotlib.org/users/gridspec.html

```python
data_scaled.shape
```

```python
def scatterplotMatrix ( sizeinch, dpival, dimensions, filename ):
    fig = plt.figure(figsize=(sizeinch, sizeinch), dpi=dpival)
    
    matrixlength = dimensions # 4-1 = 3

    gs = gridspec.GridSpec(matrixlength, matrixlength)
    
    #here comes a loop creating the matrix
    for i in range(0, matrixlength): 
        for ii in range(0, matrixlength): 
            plt.subplot(gs[i, ii]).scatter(data_scaled[:,(ii)],data_scaled[:,i], alpha=0.5, c='#262626', linewidth=0.15)    
    fig.savefig(filename)
```

```python
scatterplotMatrix (50, 400, 11, 'scatterplotmatrix_slump.pdf')
```

Caution ⇓ takes 36.771 seconds (calculated with "%prun scatterplotMatrix (50, 300, 14, 'scatterplotmatrix_winedata.pdf')"
