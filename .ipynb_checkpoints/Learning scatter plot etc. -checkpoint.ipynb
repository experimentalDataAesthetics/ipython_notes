```python
import numpy as np
import prettyplotlib as ppl
from StringIO import StringIO
# This is "import matplotlib.pyplot as plt" from the prettyplotlib library
import matplotlib.pyplot as plt# This is "import matplotlib as mpl" from the prettyplotlib library
import matplotlib as mpl
from IPython.display import HTML, Image, IFrame
#%pylab inline
```

```python
data = np.genfromtxt("iris.data", dtype=float, delimiter=',', skip_header=1) 
```

```python
data_scaled = (data - np.min(data, axis=0))/(np.max(data, axis=0) - np.min(data, axis=0))
```

```python
fig, ax = plt.subplots(1)
ppl.scatter(ax, data[:,1],data[:,0], label=str('text'), facecolor='red')
ppl.legend(ax, loc='lower right')
ax.set_title('iris test')
```

```python
fig, ax = plt.subplots(1)
ppl.scatter(ax, data[:,2],data_scaled[:,0], label=str('text'))
ppl.legend(ax, loc='lower right')
ax.set_title('iris test')
```

```python
fig, ax = plt.subplots(1)
ppl.scatter(ax, data[:,3],data[:,0], label=str('text'), facecolor='green')
ppl.legend(ax, loc='lower right')
ax.set_title('iris test')
```

## Scatterplot matrix?

```python
fig, ax = plt.subplots(1)
ppl.scatter(ax, data[:,3],data[:,0], facecolor='purple')
```

```python
def scatterplotMake ( xdim, ydim):
    fig, ax = plt.subplots(1)
    ppl.scatter(ax, data[:,xdim],data[:,ydim], facecolor='red')
```
