# EXPLORATIONS WITH ARRAY READING USING OSCGRAIN

## Wine data set (13 dimensions)

### Use code below in SC3 to load the Synthdefs, that create the grain sounds

```python
import numpy as np
from StringIO import StringIO
from IPython.display import Image
import OSC
import time, sys
import time, random
client = OSC.OSCClient()
client.connect( ( '127.0.0.1', 57110 ) )
```

```python
def oscgrain( frequency ):
    msg = OSC.OSCMessage()
    msg.setAddress("s_new")
    msg.append("grain")
    msg.append(-1)
    msg.append(0)
    msg.append(1)
    msg.append("amp")
    msg.append(1)
    msg.append("freq")
    msg.append(frequency)     #read in data points
    msg.append("sustain")
    msg.append(0.015) #0.01-0.04
    msg.append("pan")
    msg.append(-1)
    client.send(msg)
```

Playing a grain with frequency 1000Hz

```python
oscgrain(1000)
```

Load a dataset into an array

```python
data = np.genfromtxt("wine_names.csv", dtype=float, delimiter=',', skip_header=1) 
```

Explore data with .shape 

```python
data.shape
```

### Min Max Scale all data values to values between 0 and 1

```python
data_all = (data - np.min(data, axis=0))/(np.max(data, axis=0) - np.min(data, axis=0))
```

```python
data_all
```

### read-in one data point

```python
for i in range(np.asarray(data.shape)[1]-1):
    oscgrain(200+(6500*data_all[0,i+1]))
    time.sleep(0.01)
```

Use data.shape for loop settings

```python
np.asarray(data.shape)[0] # get number of datapoints
```

```python
np.asarray(data.shape)[1]-1 # get number of dimensions
```

build in these values into the loop

## and play

```python
for i in range(np.asarray(data.shape)[0]):
    for ii in range (np.asarray(data.shape)[1]-1): #first dimension is class information
        oscgrain(200+(6500*data_all[i,ii+1]))
        time.sleep(0.01) # 14*0.01=0.14 lenght/period time
    time.sleep(0.2)
```

These data are the results of a chemical analysis of wines grown in the same region in Italy but derived from three different cultivars. The analysis determined the quantities of 13 constituents found in each of the three types of wines. 

The data was used with many others for comparing various classifiers. The classes are separable, though only RDA [Regularized discriminant analysis] has achieved 100% correct classification. (RDA : 100%, QDA 99.4%, LDA 98.9%, 1NN 96.1% (z-transformed data)) (All results using the leave-one-out technique)

In a classification context, this is a well posed problem with "well behaved" class structures. A good data set for first testing of a new classifier, but not very challenging.
